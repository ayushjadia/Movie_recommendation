```python
def normalize_array(arr):
    magnitude = np.sqrt(np.sum(np.square(arr)))
    return arr / magnitude

# Apply the normalization function to each array in the column
df_2['merged_scores'] = df_2['merged_scores'].apply(normalize_array)
```

```python
indices=pd.Series(df_movies.index, index=df_movies["title"]).drop_duplicates()
indices
```

```python
def find_index(movieID):
    index_array = df_2[df_2["movieId"] == movieID].index.to_numpy()
    index_array_int = index_array.astype(int)
    return index_array_int[0]


def get_rec(title,cosine_sim_mat):
    movieID=indices[title]
    index=find_index(movieID)
    sim_score=enumerate(cosine_sim_mat[index])
    sim_score=sorted(sim_score,key=lambda x: x[1], reverse=True)
    sim_score=sim_score[1:11]
    sim_index=[i[0] for i in sim_score]
    movieIDS = df_2["movieId"].iloc[sim_index]
    list=[]
    for i,movie in enumerate(movieIDS):
        print("Reccomendations: ")
        print(f" {i+1}: {df_movies['title'][df_movies['movieId']==movie].values[0]} ")
    return list
```
